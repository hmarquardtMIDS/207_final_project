```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import os
from tqdm import tqdm
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import glob
```

```python
# Define the path to image data
image_data_path = '/Users/tylercross/git/207_hw/final_project/emotions/dataset'
# List directory contents into a DataFrame
file_names = pd.DataFrame(os.listdir(image_data_path), columns=['Files_Name'])
```

```python
# Generate a list of file paths from directories and subdirectories
file_paths = [file for file in glob.glob(image_data_path + "//*//*")]
np.random.shuffle(file_paths)
```

```python
# Extract labels from the file paths
labels = [os.path.dirname(file_path).split("/")[-1] for file_path in file_paths]
```

```python
# Create a DataFrame from files and labels
image_label_data = pd.DataFrame(zip(file_paths, labels), columns=["Image", "Label"])
# Display the DataFrame
print(image_label_data)
```

```python
# Plot label distribution using Matplotlib
label_counts = image_label_data['Label'].value_counts()
plt.bar(label_counts.index, label_counts)
plt.xticks(rotation=0)
plt.xlabel('Labels')
plt.ylabel('Frequency')
plt.title('Distribution of Labels in the Dataset')
plt.show()
```
